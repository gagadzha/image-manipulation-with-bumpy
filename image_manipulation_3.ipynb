```python
import os
from PIL import Image
import matplotlib.pyplot as plt

# function to handle alternating mirroring
def alterMirror(loc_input, width, height):
    count = 0
    img = Image.open(loc_input)
    img = img.convert("RGBA")

    for x in range(0, width):
        for y in range(0, height):
            if count % 2 == 0:
                pixel = img.getpixel((x, y))
                r, g, b, a = pixel
                img.putpixel((x, y), (r, 0, 0, a)) # setting green and blue values to zero
            else:
                pixel = img.getpixel((x, y))
                r, g, b, a = pixel
                img.putpixel((x, y), (0, 0, r, a)) # setting red value to 255 and green, blue values to zero
            count += 1
    return img

# specify the image location
loc_input = os.path.join('data', "input", "pickle_rick.png")

# get image dimensions
img = Image.open(loc_input)
width, height = img.size

# call the function and save the image
altered_img = alterMirror(loc_input, width, height)
altered_img.save(os.path.join("data", "output", "pickle_rick_altered_RGB.png"))
plt.imshow(altered_img)
plt.axis('off') # Hide the axis
plt.show()
```
