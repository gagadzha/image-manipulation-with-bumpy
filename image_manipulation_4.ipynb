```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
import os

# Assuming your script is in the root directory of your project
loc_input = os.path.join('data', 'input', "pickle_rick.png")
image = Image.open(loc_input)
np_image = np.array(image)
plt.imshow(np_image)
plt.show()
print(np_image.shape)


# Tile the image 8 times horizontally and 3 times vertically
tiled_np_image = np.tile(np_image, (3, 8, 1))

# Convert the numpy array back to a PIL image
tiled_image = Image.fromarray(tiled_np_image)

# Define the output directory and create it if it doesn't exist
output_dir = '/mnt/data/output/'
os.makedirs(output_dir, exist_ok=True)

# Save the new image
output_path = os.path.join(output_dir, 'tiled_image.png')
tiled_image.save(output_path)

# Display the tiled image
plt.imshow(tiled_image)
plt.axis('off')  # Hide the axis
plt.show()
```

```python
from PIL import Image
import numpy as np
import os

# Load the image
loc_input = os.path.join('data', 'input', "pickle_rick.png")
loc_output = os.path.join('data', 'output')
image = Image.open(loc_input)
np_image = np.array(image)

np_image.shape

# Get the size of the original image
img_height, img_width = np_image.shape[:2]

# Initialize a new array to hold the tiled image
tiled_height = img_height * 3
tiled_width = img_width * 8
tiled_image = np.zeros((tiled_height, tiled_width, np_image.shape[2]), dtype=np_image.dtype)

# Manually repeat the image by copying it into the correct positions
for i in range(3):  # Vertical replication
    for j in range(8):  # Horizontal replication
        vertical_start = i * img_height
        vertical_end = (i + 1) * img_height
        horizontal_start = j * img_width
        horizontal_end = (j + 1) * img_width
        tiled_image[vertical_start:vertical_end, horizontal_start:horizontal_end] = np_image

# Display the tiled image
plt.imshow(tiled_image)
plt.axis('off')  # Hide the axis
plt.show()

# Save the image
output_image = Image.fromarray(tiled_image)
output_image.save(os.path.join(loc_output, 'tiled_pickle_rick.png'))
```
